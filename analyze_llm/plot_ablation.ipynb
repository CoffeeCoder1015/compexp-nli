# Ablation Analysis Plotting

```python
import ablate
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os

# 1. Run the pipeline (this might take some time if not already run)
good_path, bad_path, iou_path, weight_contrib_path = ablate.run_ablation_pipeline()

# 2. Load the results
df_good = pd.read_csv(good_path)
df_bad = pd.read_csv(bad_path)
df_iou = pd.read_csv(iou_path)

# 3. Plotting function - Good vs Bad
def plot_comparison(df_good, df_bad):
    plt.figure(figsize=(10, 6))
    plt.plot(df_good['percentile'] * 100, df_good['accuracy'] * 100, marker='o', linestyle='-', label='Good Neurons')
    plt.plot(df_bad['percentile'] * 100, df_bad['accuracy'] * 100, marker='s', linestyle='--', label='Bad Neurons')
    plt.title('Comparison: Good vs Bad Neuron Cumulative Ablation Effect')
    plt.xlabel('Percentile of Neurons Ablated (%)')
    plt.ylabel('Accuracy (%)')
    plt.legend()
    plt.grid(True)
    plt.show()

plot_comparison(df_good, df_bad)

# 4. Plot combined curves (good + bad + IoU ranked)
plt.figure(figsize=(14, 8))
plt.plot(df_good['percentile'] * 100, df_good['accuracy'] * 100, 
         marker='o', linewidth=2.5, label='Good Neurons', color='green', markersize=8)
plt.plot(df_bad['percentile'] * 100, df_bad['accuracy'] * 100, 
         marker='s', linewidth=2.5, label='Bad Neurons', color='red', markersize=8)
plt.plot(df_iou['percentile'] * 100, df_iou['accuracy'] * 100, 
         marker='^', linewidth=2.5, label='IoU Ranked (All)', color='blue', markersize=8)
plt.xlabel('Percentile of Neurons Ablated (%)', fontsize=12)
plt.ylabel('Accuracy (%)', fontsize=12)
plt.title('Ablation Analysis: Good vs Bad vs IoU Ranked')
plt.legend(loc='best', fontsize=10)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
